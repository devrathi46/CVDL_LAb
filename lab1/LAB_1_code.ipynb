```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('/Users/devrathi/Desktop/CVDL_Lab/LAB_1/dog.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
#convert from BGR to RGB for proper colour displays 
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
gaussian_blur=cv2.GaussianBlur(img_rgb,(15,15),0)
```

```python
plt.imshow(gaussian_blur)
plt.show
```

```python
gaussian_blur=cv2.GaussianBlur(img_rgb,(3,3),0)
plt.imshow(gaussian_blur)
plt.show()
```

```python
#apply sharpening filter
kernel_sharpening=np.array([[0,-1,0],
                            [-1,5,-1],
                            [0,-1,0]])
sharpened=cv2.filter2D(img_rgb,-1,kernel_sharpening)
```

```python
plt.imshow(sharpened)
plt.show()
```

```python
# apply median filter
median_blur=cv2.medianBlur(img_rgb, 5)


```

```python
plt.imshow(median_blur)
plt.show()
```

```python
#average filter
# Apply a Box Filter
kernel_size = (5, 5)  
box_filtered = cv2.blur(img_rgb, kernel_size)
```

```python
plt.imshow(box_filtered)
plt.show()
```
